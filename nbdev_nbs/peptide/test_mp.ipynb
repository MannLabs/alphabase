```python
import pandas as pd
repeat = 100
peptides = ['AGHCEWQMKAADER']*repeat
mods = ['Acetyl@Protein N-term;Carbamidomethyl@C;Oxidation@M']*repeat
sites = ['0;4;8']*repeat
peptides += ['AGHCEWQMK']*repeat
mods += ['']*repeat
sites += ['']*repeat

precursor_df = pd.DataFrame({
    'sequence': peptides,
    'mods': mods,
    'mod_sites': sites
})
precursor_df['nAA'] = precursor_df['sequence'].str.len()
precursor_df['charge'] = 2
```

```python
from alphabase.peptide.precursor import calc_precursor_isotope_mp

precursor_df = calc_precursor_isotope_mp(precursor_df, processes=2, mp_batch_size=50)
precursor_df
```

```python
import numpy as np
assert np.allclose((precursor_df.isotope_mz_m1-precursor_df.precursor_mz)*precursor_df.charge, 1.0033)
```
